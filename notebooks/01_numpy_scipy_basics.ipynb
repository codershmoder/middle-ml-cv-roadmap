```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Example image
img_path = r'D:/Python things/middle-ml-cv-roadmap/data/raw/img_example.jpg'
img = cv.imread(img_path)
```

```python
if img is None:
    print("Error: Could not load image.")
else:
    # Display the image in a window named 'Example Image'
    cv.imshow('Example Image', img)
    # Wait indefinitely for a key press (0 means wait forever)
    cv.waitKey(0)
    # Destroy all created windows
    cv.destroyAllWindows()
```

```python
# Color Image Details
print(f'Array shape - {img.shape}')
print(f'Number of dimensions in array - {img.ndim}')
```

### Grayscale Conversion

```python
# Grayscale Imade Details
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv.imshow('Example Image Grayscale', img_gray)
# Wait indefinitely for a key press (0 means wait forever)
cv.waitKey(0)
# Destroy all created windows
cv.destroyAllWindows()
```

### Image Resizing

```python
# Resizing by Specifying Width and Height
new_width = 300
new_height = 200
new_dimensions = (new_width, new_height)

resized_img_1 = cv.resize(img, new_dimensions)
```

```python
cv.imshow('Original Image', img)
cv.imshow('Resized Image', resized_img_1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Resizing by Scaling Factor
sf_up = 1.13
sf_down = 0.77

resized_up = cv.resize(img, None, fx=sf_up, fy=sf_up)
resized_down = cv.resize(img, None, fx=sf_down, fy=sf_down)
```

```python
cv.imshow("Original Image", img)
cv.imshow("Scaled Up Image", resized_up)
cv.imshow("Scaled Down Image", resized_down)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Resizing with Different Interpolation Methods (for different scaling values for width (fx) and height (fy))
resized_area = cv.resize(img, None, fx=sf_down, fy=sf_up, interpolation=cv.INTER_AREA)
resized_linear = cv.resize(img, None, fx=sf_down, fy=sf_up, interpolation=cv.INTER_LINEAR)
resized_cubic = cv.resize(img, None, fx=sf_down, fy=sf_up, interpolation=cv.INTER_CUBIC)
resized_nearest = cv.resize(img, None, fx=sf_down, fy=sf_up, interpolation=cv.INTER_NEAREST)
```

```python
cv.imshow("Original Image", img)
cv.imshow("Resized with INTER_AREA", resized_area)
cv.imshow("Resized with INTER_LINEAR", resized_linear)
cv.imshow("Resized with INTER_CUBIC", resized_cubic)
cv.imshow("Resized with INTER_NEAREST", resized_nearest)
cv.waitKey(0)
cv.destroyAllWindows()
```
